1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

```python
# pd.read_csv('Laliga.csv')  
# since the header is also considered as a data row, 'header' parameter is to be used to define the same.

laliga_df = pd.read_csv('Laliga.csv',header=1)
```

4. Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 

```python
# function to return all the teams with their Goal Differences
def Goal_diff_count(df,fromCol,ToCol,diff):
    df[diff] = df[fromCol] - df[ToCol]
    return df
result = Goal_diff_count(laliga_df,'GoalsFor','GoalsAgainst','goal_diff')
```

```python
#print(max(laliga_df['goal_diff']))
#print(min(laliga_df['goal_diff']))
print("The team with maximum goal difference of",max(laliga_df['goal_diff']),"is",laliga_df.sort_values(by='goal_diff',ascending=True).head(1)['Team'].values[0])
print("The team with minimum goal difference of",min(laliga_df['goal_diff']),"is",laliga_df.sort_values(by='goal_diff',ascending=False).head(1)['Team'].values[0])
```

5. Create a new column with name “Winning Percent” and append it to the data set

Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
laliga_df['Percentage of Winning'] = laliga_df['GamesWon']/laliga_df['GamesPlayed']*100
```

```python
laliga_df['Percentage of Winning'].fillna(0,inplace=True)
laliga_df['Percentage of Winning']
print("The top 5 teams with highest Winning percentage are")
for i in range(0,5):
    print(laliga_df.sort_values(by='Percentage of Winning',ascending=False).head(5)['Team'][i],"with",round(laliga_df.sort_values(by='Percentage of Winning',ascending=False).head(5)['Percentage of Winning'][i],1),"%")
```
